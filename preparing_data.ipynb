```python
# data cleaning and text modification
other=df.drop(columns=['Cabin'])
other['Age']=other['Age'].fillna(other['Age'].mean())
other['Embarked']=other['Embarked'].fillna(other['Embarked'].mode()[0])
other['Survived']=other['Survived'].astype(str).str.replace('0','un_survived=')
other['Survived']=other['Survived'].astype(str).str.replace('1','survived')
other=other.sort_values(by=['Age','Fare'])
other['total_family']=other['SibSp']+other['Parch']+1
other['Pclass']=other['Pclass'].astype(str)
other['Pclass']=other['Pclass'].str.replace('1','first_class')
other['Pclass']=other['Pclass'].str.replace('2','second_class')
other['Pclass']=other['Pclass'].str.replace('3','third_class')


```

```python
# saving data
other.to_csv(r'C:\Users\Paragon\OneDrive\Desktop\titanic_data\cleaned-Titanic-Dataset.csv')
```
